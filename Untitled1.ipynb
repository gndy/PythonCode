```python
from Bio import SeqIO
import os


def get_positions(filename):
    mdict = {}
    seqs = SeqIO.parse(mdir_fgene+filename,'fasta')
    for seq in seqs:
        header_items = seq.description.split("_")
        name = '_'.join(header_items[0:2])
        strand_symbol = header_items[-1]
        if strand_symbol == '+':
            strand = 1
        else:
            strand = -1
        mdict[name] = (int(header_items[2])-1,int(header_items[4]),strand)
        
    return mdict

def isOverlap(t1,t2):
    tmp = [t1,t2]
    tmp.sort()
    if tmp[1][0]<tmp[0][1]:
        return True
    else:    
        return False
    
def overlapCount(t1,t2):
    tmp = [t1,t2]
    tmp.sort()
    if tmp[1][1]<tmp[0][1]:
        count = tmp[1][1]-tmp[1][0]
    else:
        count = tmp[0][1]-tmp[1][0]
    return count
    

def cmp_positions(fgene,origins):
    
    
    total_length = 0
    for origin in origins:
        pos_origin = (origin.start.position,origin.end.position,origin.strand)
        if fgene[2] == pos_origin[2] and isOverlap(fgene[0:2],pos_origin[0:2]):
            overlap_length = overlapCount(fgene[0:2],pos_origin[0:2])
            total_length += overlap_length
            
    return total_length
        

                
    

mdir = "/home/wenlei/LUO/Simon/non-humans_gb/"
mdir_fgene = "/home/wenlei/LUO/Simon/2/"
mout_dir = "/home/wenlei/LUO/Simon/overlap_results/"



for mfile in os.listdir(mdir):
    mdict = get_positions(mfile)
    cds_locs = []
    mdict_2 = {}

    record = SeqIO.read(mdir+mfile,'genbank')
    print mfile
    for feature in record.features:
        mlocation = feature.location
        if feature.type == 'CDS':
            cds_loc = feature.location.parts
            cds_locs.append(cds_loc)
            
    for (k,v) in mdict.items():
        last_list = [v]
        for item in cds_locs:
            moverlap_length = cmp_positions(v,item)
            if  moverlap_length > 0:
                tmplist = [(origin.start.position,origin.end.position,origin.strand) for origin in item]
                tmplist.append(moverlap_length)
                last_list.append(tmplist)

        if len(last_list) != 1:
            mdict_2[k] = last_list
    #print mdict_2
    
    key_list = []
    writing_list = []
    
    for (k,v) in mdict_2.items():
        if len(v) == 2 and v[0] in [ kk[0:3] for kk in v[1][0:-1]]:
            #print "right"
            pass
        else:
            key_list.append(int(k.split('_')[-1]))
            
    key_list.sort()
    
    for mkey in key_list:
        mkey = "GENE_"+str(mkey)
        writing_list.append(mkey)
        mvalue = mdict_2[mkey]
        writing_list.append(str(mvalue[0]))
        for jjj in mvalue[1::]:
            writing_list.append('-'.join([str(kkk) for kkk in jjj]))
            
            
    out_content = '\n'.join(writing_list)
    mwriting_handle = open(mout_dir+mfile,'w')
    mwriting_handle.write(out_content)
    mwriting_handle.close()
        
    


                
            
        


    
    
```

```python
import os,time
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

ori_dir = "/home/wenlei/LUO/Simon/cmv_cdss/"
new_gene_dir = "/home/wenlei/LUO/Simon/hits_dir/fgenev/output_removed_with_protein_fullname/"
raw_fasta_dir = "/home/wenlei/LUO/Simon/raw_fasta/"

filenames_new = os.listdir(new_gene_dir)

st_dict = {'-':'-1','+':'1'}

for filename in filenames_new:
    print filename
    total_seqs = []
    
    content_1 = SeqIO.parse(ori_dir+filename,'fasta')#original annotation
    for i in content_1:
        seq_description = i.description.split('|')
        total_seqs.append(i)
    gn = seq_description[11].strip()#gn
    tx = seq_description[-2].strip()#taxonomy

    
    protein_dict = {}
    content_2 = SeqIO.parse(raw_fasta_dir+filename,'fasta')#raw fasta for protein seq
    for j in content_2:
        raw_description = j.description
        protein = j.seq
        protein_dict[raw_description] = protein

    
    content = open(new_gene_dir+filename).read().splitlines()
    content = [i for i in content if i.strip()]
    time_base = int(str(time.time())[3:10])
    
    for ii in content:
        gene_name = ii.split('\t')[0]
        product = ii.split('\t')[1]
        lo = (int(gene_name.split('_')[2]),int(gene_name.split('_')[4]))#location
    
    
    
    for ii in content:
        gene_name = ii.split('\t')[0]
        product = ii.split('\t')[1]+'_fgenesv'
        lo = gene_name.split('_')[2]+'-'+gene_name.split('_')[4]#location
        st = st_dict[gene_name.split('_')[-1]]#strand
        product_seq = protein_dict[gene_name]
        time_base += 1
        gi = '000'+str(time_base)
        gb = "IOV"+str(time_base)
        new_seq_description = 'gi|'+gi+'|gb|'+gb+'|or|1'+'|st|'+st+'|lo|'+lo+'|gn|'+gn+'|tx|'+tx+'|'+product
        mseqrecord = SeqRecord(product_seq,id=new_seq_description,description='')
        total_seqs.append(mseqrecord)
    print total_seqs
    SeqIO.write(total_seqs,'/home/wenlei/tmp/'+filename,'fasta')
    
    
    
        
        
```

```python
from Bio import SeqIO
import os,re

dir_1 = "/home/wenlei/tmp/"
dir_2 = "/home/wenlei/tmp2/"

filenames = os.listdir(dir_1)
for filename in filenames:
    print filename
    seq_dict = {}
    seqs = SeqIO.parse(dir_1+filename,'fasta')
    for seq in seqs:
        description = seq.description
        lo = description.split('|')[9]
        lo = (int(lo.split('-')[0]),int(lo.split('-')[1]))
        seq_dict[lo] = seq
    lo_list = seq_dict.keys()
    lo_list.sort()
    for i in range(len(lo_list)):
        mkey = lo_list[i]
        replacement = seq_dict[mkey].description
        replacement = re.sub('or\|\d{1,3}','or|'+str(i+1),replacement)
        seq_dict[mkey].id = replacement
        seq_dict[mkey].description = ''
        
    output_list = []
    
    for i in lo_list:
        output_list.append(seq_dict[i])

    SeqIO.write(output_list,dir_2+filename,'fasta')
    
        

        
```
